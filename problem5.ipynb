```python
import pandas as pd

df = pd.read_csv("kc_house_data.csv")

# Drop ignored columns
df = df.drop(columns=["id", "date", "zipcode"])

price = df["price"]
features = df.drop(columns=["price"])

stats = pd.DataFrame({
    "Average": features.mean(),
    "Min": features.min(),
    "Max": features.max(),
    "Variance": features.var()
})

stats
```

The feature with the lowest average is waterfront, with an average value of approximately 0.0075, meaning that very few houses are waterfront properties.

The feature with the highest average is sqft_living, with an average value of approximately 2079.9 square feet.

```python
correlations = features.corrwith(price).sort_values(ascending=False)
correlations
```

All features have a positive correlation coefficient with the response variable (price), meaning that increases in these features are generally associated with higher house prices.

The feature with the highest positive correlation with price is sqft_living, with a correlation coefficient of ~0.70, indicating a strong positive relationship between living area and house price.

There were no features with a negative correlation coefficient with the response variable (price). All features showed positive correlation coefficients, indicating that increases in these features are generally associated with higher house prices.
