```python
import numpy as np

np.random.seed(4400)

matrices = [np.random.randint(-10, 11, size=(3,3)) for _ in range(3)]
```

```python
for i, M in enumerate(matrices, start=1):
    print(f"\nMatrix {i}:")
    print(M)
    
    try:
        M_inv = np.linalg.inv(M)
        print("\nInverse:")
        print(M_inv)
        
        print("\nVerification (M × M⁻¹):")
        print(M @ M_inv)
        
    except np.linalg.LinAlgError:
        print("\nThis matrix is not invertible (determinant = 0).")
```

Each matrix was generated randomly. NumPy's `linalg.inv` function was used to attempt
to compute the inverse. When the inverse exists, multiplying the matrix
by its inverse yields the identity matrix (up to floating-point error).
If the matrix is singular, NumPy throws an error, indicating that the
inverse does not exist.
